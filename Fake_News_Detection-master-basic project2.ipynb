```python
import re
import string

def wordopt(news):
    news = news.lower()
    news = re.sub(r'\[.*?\]', '', news)  # Remove text within square brackets
    news = re.sub(r'\W', ' ', news)  # Remove non-word characters (keep spaces)
    news = re.sub(r'http://\s+|www\.\s+', '', news)  # Remove URLs
    news = re.sub(r'<.*?>', '', news)  # Remove HTML tags
    news = re.sub(r'[%s]' % re.escape(string.punctuation), '', news)  # Remove punctuation
    news = re.sub(r'\n', ' ', news)  # Remove newline characters
    news = re.sub(r'\w*\d\w*', '', news)  # Remove words containing digits
    news = re.sub(r'\s+', ' ', news)  # Replace multiple spaces with a single space
    news = news.strip()  # Remove leading and trailing spaces
    return news
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
x=data['text']
y=data['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create an instance of TfidfVectorizer
vectorization = TfidfVectorizer()

# Fit and transform the training data
xv_train = vectorization.fit_transform(x_train)

# Transform the test data
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
# Create an instance of LogisticRegression
LR = LogisticRegression()
# Fit the model on the training data
LR.fit(xv_train, y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not Fake News"
    else:
        return "Unknown"

def manual_testing(news):
    # Ensure you have the function 'wordopt' defined somewhere above
    # Ensure you have the models 'LR', 'DT', 'GBC', 'RFC' defined and trained somewhere above

    # Prepare the test data
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    
    # Transform the text data using the same vectorizer used for training
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    
    # Get predictions from each model
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GB = GB.predict(new_xv_test)
    pred_RF = RF.predict(new_xv_test)
    
    # Print the predictions
    print("\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(
        output_label(pred_LR[0]),
        output_label(pred_DT[0]),
        output_label(pred_GB[0]),
        output_label(pred_RF[0])
    ))


```
